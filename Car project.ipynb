#### Show distribution of categorical columns.

```python
print(df['Brand'].value_counts())
print(df['Color'].value_counts())
print(df['Location'].value_counts())
print(df['Model'].value_counts())
```

## Graphs to show Numerical data:

#### Histogram showing data of distribution of car over years.

```python
df['Year'].hist(bins=20, color="g")
plt.title('Distribution(production) of Cars over Years')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
```

#### Histogram showing data of car mileage. 

```python
df['Mileage'].hist(bins=20, color="red")
plt.title('Distribution of Car Mileage')
plt.xlabel('Mileage')
plt.ylabel('Frequency')
plt.show()
```

#### Histogram showing data of distribution of car price. 

```python
df['Price'].hist(bins=20, color="b")
plt.title('Distribution of Car Price(range of price of cars)')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

#### Showing outliers(different values) in numerical columns. 

```python
import seaborn as sns

sns.boxplot(data=df[['Year', 'Mileage', 'Price']])
plt.title('Box Plots for Numerical Columns')
plt.show()
```

## Graphs showing Categorical data:

#### Bar chart showing distribution of brands. 

```python
df['Brand'].value_counts().plot(kind='bar', color="r")
plt.title('Frequency of Brands')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.show()
```

#### Bar chart showing distribution of model.  

```python
plt.figure(figsize=(15, 8))  # Adjust the size as needed
df['Model'].value_counts().plot(kind='bar')
plt.title('Frequency of Model')
plt.xlabel('Model')
plt.ylabel('Count')
plt.xticks(rotation=90)  # Rotate x-axis labels by 90 degrees
plt.show()
```

#### Bar chart showing distribution of colors.  

```python
df['Color'].value_counts().plot(kind='bar', color="g")
plt.title('Frequency of Colors')
plt.xlabel('Color')
plt.ylabel('Count')
plt.show()
```

#### Bar chart showing distribution of Location. 

```python
df['Location'].value_counts().plot(kind='bar', color= "purple")
plt.title('Frequency of Locations')
plt.xlabel('Location')
plt.ylabel('Count')
plt.show()
```

## Graphs showing comparisons and relations:

#### Graph showing effect of mileage on price.

```python
sns.barplot(x='Mileage', y='Price', data=df)
plt.title('Mileage vs Price')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```

#### Graph showing  price over years.

```python
sns.barplot(x='Year', y='Price', data=df)
plt.title('Year vs Price')
plt.xlabel('Year')
plt.ylabel('Price')
plt.show()
```

#### Pair plots of 'year', 'price', 'milegae'.

```python
sns.pairplot(df[['Year', 'Mileage', 'Price']])
plt.show()
```
